```python
import matplotlib.pyplot as plt

# Daten
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Diagramm erstellen
plt.plot(x, y, label='x vs. y', color='blue', linestyle='-', marker='o')

# Titel und Achsenbeschriftungen
plt.title('Einfaches Diagramm')
plt.xlabel('X-Achse')
plt.ylabel('Y-Achse')

# Legende
plt.legend()

# Diagramm anzeigen
plt.show()
```
